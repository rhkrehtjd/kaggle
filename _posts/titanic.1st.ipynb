> ### 각 칼럼에 null data가 몇개 있는지 퍼센트를 보기 위한 

- {0:<10} => {0} 값을 ":<10" 10자리로 표현할건데 왼쪽 정렬
- {1:>5} =>  {1} 값을 ":>5" 5자리로 표현할건데 오른쪽 정렬
- :^10 = 가운데 정렬
- \t 이건 탭 한 번 눌러주는 기능 => 그래서 중간에 크게 띄어쓰기 가능해짐

```python
for col in df_train.columns:
    msg='column:  {:^11}\t Percent of NaN value {:.2f}%'\
    .format(col, 100*(df_train[col].isnull().sum()/df_train[col].shape[0] ))
    print(msg)
```

```python
for col in df_test.columns:
    msg='column:{:>10}\t Percent of NaN value {:.2f}%'\
    .format(col,100*(df_test[col].isnull().sum()/df_test[col].shape[0] ))
    print(msg)
```

> 위 셀에서 for문 실행하면서 가장 마지막에 col에  Embarked가 저장되면서 col이  Embarked로 끝나게 됨

> ### `Pclass`

```python
df_train[['Pclass','Survived']].groupby(['Pclass'],as_index=True).count()
```

count  $VS$  sum
- sum은 숫자 자체 데이터를 더한 것
- count는 각각 객체가 몇개 인가?

```python
pd.crosstab(df_train['Pclass'],df_train['Survived'],margins=True)\
.style.background_gradient(cmap='Greens')
```

margins => 각각 합계

index여부에 따라 plot의 형태가 달라질 수 있음 

```python
df_train[['Pclass','Survived']].groupby(['Pclass'], as_index=True)\
.mean().sort_values(by='Survived',ascending=False).plot.bar()
```

```python
y_positions =1 #title의 위치
f,ax = plt.subplots(1,2,figsize=(18,8))
df_train['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_title('Number of passengers By Pclass', y=y_positions)
ax[0].set_ylabel('Count')
sns.countplot('Pclass',hue='Survived',data=df_train, ax=ax[1])
ax[1].set_title('Pclass: Survived vs Dead', y=y_positions)
plt.show()
```
