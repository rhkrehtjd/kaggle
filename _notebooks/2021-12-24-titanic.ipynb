```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)

import missingno as msno 
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
df_train=pd.read_csv('C:/Users/ehfus/Downloads/titanic/train.csv')
df_test=pd.read_csv('C:/Users/ehfus/Downloads/titanic/test.csv')
```

> ### `Pclass`

```python
df_train[['Pclass','Survived']].groupby(['Pclass'],as_index=True).count()
```

count  $VS$  sum
- sum은 숫자 자체 데이터를 더한 것
- count는 각각 객체가 몇개 인가?

```python
pd.crosstab(df_train['Pclass'],df_train['Survived'],margins=True)\
.style.background_gradient(cmap='Greens')
```

margins => 각각 합계

index여부에 따라 plot의 형태가 달라질 수 있음 

```python
df_train[['Pclass','Survived']].groupby(['Pclass'], as_index=True)\
.mean().sort_values(by='Survived',ascending=False).plot.bar()
```

```python
y_positions =1 #title의 위치
f,ax = plt.subplots(1,2,figsize=(18,8))
df_train['Pclass'].value_counts().plot.bar(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_title('Number of passengers By Pclass', y=y_positions)
ax[0].set_ylabel('Count')
sns.countplot('Pclass',hue='Survived',data=df_train, ax=ax[1])
ax[1].set_title('Pclass: Survived vs Dead', y=y_positions)
plt.show()
```

> `SEX`

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df_train[['Sex','Survived']].groupby(['Sex'],as_index=True).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot('Sex',hue='Survived',data=df_train,ax=ax[1])
ax[1].set_title('Sex: survived vs Dead')
plt.show()
```
